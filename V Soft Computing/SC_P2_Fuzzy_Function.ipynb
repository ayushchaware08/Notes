### 1. Triangular Membership Function

```python
x = np.arange(0, 11, 1)
```

```python
#  inbuild Function
trimf = fuzz.trimf(x, [2, 5, 8])

# Plot the membership function
plt.plot(x, trimf)
plt.xlabel('x')
plt.ylabel('Membership degree')
plt.title('Triangular Membership Function')
plt.show()
```

```python
def triangular_mf(x, a, b, c):
  return np.maximum(0, np.minimum((x-a)/(b-a), (c-x )/(c-b)))
```

```python
# define range of value of x
x = np.linspace(-5, 5, 500)

# parameters
a, b, c = -3 , 0 , 3

# compute membership degrees
membership_values = triangular_mf(x, a, b, c)
```

```python
# plot
plt.figure(figsize = (8,6))
plt.plot(x, membership_values, label = 'triangular membership function')
plt.title("Triangular Membership Function")
plt.xlabel('X')
plt.ylabel('Membership Degree')
plt.show()
```

```python
# Predict Membership function
new_x = int(input("Enter the value of x: "))
predicted_mf = triangular_mf(new_x, a, b, c)
print(predicted_mf)
```

### 2. Trapazoidal Membership Function

```python
def trapezoidal_mf(x, a, b, c, d):
  return np.maximum(0,np.minimum(np.minimum((x-a)/(b-a),1),(d-x)/(d-c)))
```

```python
# define range of value of x
x = np.linspace(-5, 5, 500)

# parameters
a, b, c, d = -3 , -2 , 2, 3

# compute membership degrees
membership_values = trapezoidal_mf(x, a, b, c, d)
```

```python
# plot
plt.figure(figsize = (8,6))
plt.plot(x, membership_values, label = 'Trapazoidal membership function')
plt.title("Trapazoidal Membership Function")
plt.xlabel('X')
plt.ylabel('Membership Degree')
plt.show()
```

```python
# Predict Membership function
new_x = int(input("Enter the value of x: "))
predicted_mf = trapezoidal_mf(new_x, a, b, c, d)
print(predicted_mf)
```

### 3. Bell Membership Function

```python
def bell_mf(x,a,b,c):
  return 1/ (1+ np.abs((x-c)/a)**(2*b))
```

```python
# define range of value of x
x = np.linspace(-5, 5, 500)

# parameters
a, b, c = 1,2,0

# compute membership degrees
membership_values = bell_mf(x, a, b, c)
```

```python
# plot
plt.figure(figsize = (8,6))
plt.plot(x, membership_values, label = 'Bell membership function')
plt.title("bell Membership Function")
plt.xlabel('X')
plt.ylabel('Membership Degree')
plt.show()
```

```python
# Predict Membership function
new_x = int(input("Enter the value of x: "))
predicted_mf = bell_mf(new_x, a, b, c)
print(predicted_mf)
```
